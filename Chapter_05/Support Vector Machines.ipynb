```python
# The following code loads the iris dataset and the goal of it is to identiy which flowers are virginica Flowers
```

```python
from sklearn.datasets import load_iris
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import LinearSVC
```

```python
iris = load_iris(as_frame=True)
X = iris.data[["petal length (cm)", "petal width (cm)"]].values
y = (iris.target == 2) # Iris viginica
```

```python
svm_clf = make_pipeline(StandardScaler(),
                        LinearSVC(C=1, random_state=42))
svm_clf.fit(X, y)
```

```python
# Lets use the model to make predictions
X_new = [[5.5, 1.7] , [5.0, 1.5]]
svm_clf.predict(X_new)
```

```python
# Lets look at the score 
svm_clf.decision_function(X_new)
```

```python
# Nonlinear SVM classification

# Lets create a pipleline that contains a polynomialFeature transformer so we can identiy claasses that are in a linear fasion

from sklearn.datasets import make_moons
from sklearn.preprocessing import PolynomialFeatures

X, y = make_moons(n_samples=100, noise=0.15, random_state=42)

Polynomia_svm_clf = make_pipeline(
    PolynomialFeatures(degree=3),
    StandardScaler(), 
    LinearSVC(C=10, max_iter=10_000, random_state=42)
)
Polynomia_svm_clf.fit(X, y)
```

```python
# Polynomial Kernal 
## This give us the ability to have a high degree polynomial without the time cost of adding the high degree. In other words, the tradition way of having a high degree polynomial would take a very long time.
## By utilizing the Polynomial kernal we can still get the benifet with of a high degree without the time cost 

from sklearn.svm import SVC
poly_kernel_svm_clf = make_pipeline(StandardScaler(),
                                    SVC(kernel='poly', degree=3, coef0=1, C=5))
poly_kernel_svm_clf.fit(X, y)
```

```python
# Gausian RBF (Similarity features)

rbf_kernal_svm_clf = make_pipeline(StandardScaler(),
                                   SVC(kernel='rbf', gamma=5, C=0.001)) # Kernal - helps utilize other models methods without the computations cost
# kernal rbf is like using Gaussian model
# gamma sets the y value and makes the bell curve more narrow(high value) or wide (low value)
# C - sets how wide the 'street' should be to split the data points
rbf_kernal_svm_clf.fit(X, y)
```

```python
# Train a svm classifer on the wine datset. This Dataset contains the chemical analysis of 178 wine samples produced by 3 different cultivators: the goal 
# is to train a classification model capable of predicting the cultivator based on the wines chemical analysis. Since the SVM classifiers are binary classifiers
# we will need to use one-versus-all to classify all three classes. What accuracy can I reach?

```

```python
from sklearn.datasets import load_wine

wine = load_wine(as_frame=True)
```

```python
print(wine.DESCR)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    wine.data, wine.target, random_state=42)
```

```python
import seaborn as sns
```

```python
# First we will do a linear SVM classification
# remember we will have to use a one-versus-all

# lets import the libraries 
from sklearn.pipeline import make_pipeline # Helps us stream line the transformation of the dataset
from sklearn.preprocessing import StandardScaler # Helps make sure all the values are on the same scale 
from sklearn.svm import LinearSVC # The model we will be using first
from sklearn.metrics import accuracy_score
```

```python
lin_svm = LinearSVC()
# linear SVC uses one-versus-all by default
svm_clf = make_pipeline(StandardScaler(),
                        LinearSVC(C=1))
```

```python
svm_clf.fit(X_train, y_train)
```

```python
svm_clf.score(X_test, y_test)
```

```python
y_pred = svm_clf.predict(X_test)
```

```python
# evaluate the score 
accuracy_score(y_test, y_pred)
```

```python
# let's make the data into a polynimial to the 3rd degree
poly_cvm_clf = make_pipeline(
    PolynomialFeatures(degree=3),
    StandardScaler(),
    LinearSVC(C=1)
)
```

```python
poly_cvm_clf.fit(X_train, y_train)
```

```python
y_pred = poly_cvm_clf.predict(X_test)
```

```python
accuracy_score(y_test, y_pred)
```

```python
# Gaussian RBF Kernal
rbf_kernal_svm_clf = make_pipeline(StandardScaler(),
                                   SVC(kernel='rbf', gamma=1, C=20))
rbf_kernal_svm_clf.fit(X_train, y_train)
```

```python
y_pred = rbf_kernal_svm_clf.predict(X_test)
```

```python
accuracy_score(y_test, y_pred)
```
